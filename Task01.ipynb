```python
#Task 01(Classification)
#Dataset: Breast Canser.csv
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset

df = pd.read_csv("C:/DS/Final_assignment(36508)/1.breast-cancer.csv")
df.head()
```

```python
#Understand the Dataset

df.info()
df.describe()
df.isnull().sum()
```

The dataset(Breast_cancer dataset) contains features computed from breast cancer cell images. 

The traget variable is diagnosis, which is categorical.

```python
#Data Cleaning and Preprocessing
#Drop Unnecessary columns

df.drop(['id'], axis=1, inplace=True)
```

```python
#Encode Target Variables(diagnosis)

from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
df['diagnosis']= le.fit_transform(df['diagnosis'])
```

```python
#exploratory Data Analysis(EDA)
#Target Variable Distribution

sns.countplot(x='diagnosis', data=df)
plt.title("Diagnosis Distribution (0= Benign, 1= Malignant)")
plt.show()
```

```python
#Features Distribution

sns.histplot(df['radius_mean'], kde=True)
plt.title("Distribution of Radius Mean")
plt.show()
```

```python
#Box plot
sns.boxplot(x='diagnosis', y='area_mean', data=df)
plt.title("Area Mean Vs Diagnosis")
plt.show()
```

```python
#Correlation Heatmap

plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Malignant tumors generally show higher radius and area values compared to benign tumors.

```python
#Split Features and Target

x = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
#Train-Test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
#Build Classification Model 
#Logistic Regression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)
model.fit(x_train, y_train)
```

```python
#Make Prediction

y_pred = model.predict(x_test)
```

```python
#Model Evaluation 

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#Interpretation of Results
```

The Logistic Regression model achieved high accuracy, indicating strong performance in classifying breast cancer tumors.

The confusion matrix shows a low number of false negatives, which is critical for medical diagnosis.

```python
#Conclusion
```

This classification model successfully distinguishes between malignant and benign breast cancer tumors.

Logistic Regression proved effective due to the linear separability of the features.

The model can assist in early diagnosis, although further improvements can be made using advanced classifiers.
